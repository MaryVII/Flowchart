## Dealing With Missing Values

```python
#Drop missing values from "Head of PPC Approval"
df = df.dropna(subset=['Head of PPC Approval'])
df.head()
```

```python
df.shape
```

762 rows left after removing missing values from "`Head of PPC Approval`" column

## Dealing with duplicates

```python
# Duplicate the 'Title' and 'Created' columns
df['Title_Duplicate'] = df['Title']
df['Created_Duplicate'] = df['Created']

# Merge the duplicated columns into a new column (e.g., 'Merged_Title_Created')
df['Merged_Title_Created'] = df['Title'].astype(str) + ' - ' + df['Created'].astype(str)

# Display the updated DataFrame (Optional)
df.head()
```

```python
# Remove duplicates
df = df.drop_duplicates(subset=['Merged_Title_Created'])
df.shape
```

```python
# Recheck for duplicates
df['Title_duplicate'] = df.duplicated(subset=['Title'], keep=False)
df_true = df[df['Title_duplicate'] == True]
df_true.shape
```

```python
df_true.head(9)
```

8 Duplicate titles with different "`Created`" date identify

```python
# Drop Merged_Title_Created, Created_Duplicate
df = df.drop(['Merged_Title_Created','Created_Duplicate','Title_Duplicate'], axis=1)
df_true = df_true.drop(['Merged_Title_Created','Created_Duplicate','Title_Duplicate'], axis=1)
```

```python
# Attached signed agreement ='0.0'

df_no_attached = df[df['Attached signed agreement'] == 0.0]
df_no_attached.head()
```

```python
df_no_attached.shape
```

261 data have missing attachments

```python
#convert to excel worksheet

with pd.ExcelWriter('output.xlsx') as writer:
  df.to_excel(writer, sheet_name='Main', index=False)
  df_true.to_excel(writer, sheet_name='Duplicates', index=False)
  df_no_attached.to_excel(writer, sheet_name='Missing Attachments', index=False)

from google.colab import files
files.download('output.xlsx')
```
