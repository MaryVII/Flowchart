### Data Cleaning

```python
#Remove CapsLock
def transform_capitalized_each_words(text):
  if isinstance(text, str):
    return ' '.join(word.capitalize() for word in text.split())
  return text

df['Traveler'] = df['Traveler'].apply(transform_capitalized_each_words)
df['Travel Processor'] = df['Travel Processor'].apply(transform_capitalized_each_words)
```

```python
#Create Unique ID

df['Unique ID'] = df['Traveler'] + df['Date From'].astype(str) + df['Date To'].astype(str) + df['Created by'].astype(str)
```

```python
# Check for duplicates

df['Potential Duplicates'] = df.duplicated(subset=['Unique ID'], keep=False)
df['Potential Duplicates'] = df['Potential Duplicates'].replace({True: 'Yes', False: 'No'})
```

```python
yes_count = df['Potential Duplicates'].value_counts().get('Yes', 0)
print(f"Number of Duplicates: {yes_count}")
```

```python
df5 = df[df['Potential Duplicates'] == 'Yes']
df5.head()
```

```python
#Calculate travel duration
from datetime import datetime

def calculate_duration(row):
  try:
    date_from = row['Date From']
    date_to = row['Date To']
    if pd.isnull(date_from) or pd.isnull(date_to):
      return np.nan

    date_from = pd.to_datetime(date_from).date()
    date_to = pd.to_datetime(date_to).date()
    duration = (date_to - date_from).days + 1
    return duration
  except Exception as e:
    print(f"Error calculating duration: {e}")
    return np.nan


df['Duration'] = df.apply(calculate_duration, axis=1)
df['Duration'] = df['Duration'].fillna(0).astype(int)
```

```python
#Extract countries name from description column
import pycountry

def extract_country(description):
  if isinstance(description, str):
    for country in pycountry.countries:
      if country.name in description:
        return country.name
  return np.nan

df['Location'] = df['Short description'].apply(extract_country)
```

```python
#Clean Invoice Number
df1['Invoice Number'] = df1['Invoice Number'].str[:10]
df2['Invoice Number'] = df2['Invoice Number'].str[:10]
```
